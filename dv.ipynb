## 1. Create a histogram to analyze the distribution of Age

```python
plt.figure(figsize=(8,5))

sns.histplot(df['Age'], bins=10, kde=True)

plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.grid(True)

plt.show()
```

## 2. Plot a pie chart showing the percentage distribution of Gender.

```python
gender_counts = df['Gender'].value_counts()
gender_counts
```

```python
plt.figure(figsize=(6,6))

plt.pie(gender_counts,
        labels=gender_counts.index,
        autopct='%1.1f%%',
        startangle=90)

plt.title("Gender Distribution")
plt.show()
```

## 3. Create a bar chart showing the number of customers in each City

```python
city_counts = df['City'].value_counts().sort_values(ascending=False)
```

```python
plt.figure(figsize=(8,5))                 
plt.bar(city_counts.index,
        city_counts.values,
        color='skyblue',                 
        edgecolor='black')              
plt.title("Customers in Each City", fontsize=14)
plt.xlabel("City", fontsize=12)
plt.ylabel("Number of Customers", fontsize=12)
plt.show()
```

## 4. Plot a box plot to analyze Monthly Income distribution.

```python
plt.figure(figsize=(8,5))
sns.boxplot(
    y=df['Monthly_Income'],
    color='lightgreen'
)
plt.title("Monthly Income Distribution", fontsize=14)
plt.ylabel("Monthly Income", fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()
```

## 5. Create a histogram of Spending Score and interpret the skewness.

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Spending_Score'], 
             bins=10, 
             color='skyblue', 
             kde=True)
plt.title("Spending Score Distribution", fontsize=14)
plt.xlabel("Spending Score", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()
```

## 6. Plot a line chart to visualize Purchase Amount trend over Purchase Date.

```python
df = df.sort_values('Purchase_Date')
```

```python
df["Month"] = df["Purchase_Date"].dt.month
monthly_sales = df.groupby("Month")["Purchase_Amount"].sum()
plt.figure(figsize=(8,5))
monthly_sales.plot(color='red',marker="o",linewidth=2)
plt.title("Monthly Purchase Trend")
plt.xlabel("Month")
plt.ylabel("Purchase Amount")
plt.grid(True)
plt.show()
```

## 7. Create a bar plot to compare average Monthly Income by Gender.

```python
avg_income = df.groupby('Gender')['Monthly_Income'].mean()
```

```python
plt.figure(figsize=(6,5))
plt.bar(avg_income.index,
        avg_income.values,
        color='lightblue',
        edgecolor='black')
plt.title("Average Monthly Income by Gender")
plt.xlabel("Gender", fontsize=12)
plt.ylabel("Average Monthly Income", fontsize=12)
plt.show()
```

## 8. Plot a box plot of Purchase Amount grouped by Product Category

```python
plt.figure(figsize=(10,5))
sns.boxplot(
    x='Product_Category',
    y='Purchase_Amount',
    data=df,
)
plt.title("Purchase Amount by Product Category", fontsize=14)
plt.xlabel("Product Category", fontsize=12)
plt.ylabel("Purchase Amount", fontsize=12)
plt.grid()
plt.show()
```

## 9. Create a pie chart to show Product Category distribution

## 12. Plot a line chart showing total Purchase Amount per month

## 14. Plot a histogram to analyze Age distribution by Gender
